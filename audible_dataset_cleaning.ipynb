```python
import pandas as pd
```

```python
dt = pd.read_csv('audible_uncleaned.csv')
dt
```

```python
#we verified if it exist isnull values in dataset
dt.isnull().sum()
```

```python
#Now we will arrange columns like author, narrator, time and stars. We start with author column. 
#We will remove Writtenby text
dt['author'] = dt['author'].str.replace('Writtenby:','')
dt['author']
```

```python
#We remove Narreted by in the culomn narrator
dt['narrator'] = dt['narrator'].str.replace('Narratedby:','')
dt['narrator']
```

```python
#We treat well time column to be best written
# extract hours and minutes by using regular expression which consider space between number and texte
hours = dt['time'].str.extract(r'(\d+)\s*hrs').astype(float)  # Use float for manage Nan values
minutes = dt['time'].str.extract(r'(\d+)\s*mins').astype(float)

# Convert hours in minutes
hours_in_min = hours * 60

# Calculate sum of minutes
total_min = hours_in_min.fillna(0) + minutes.fillna(0)  # Use fillna to replace NaN by 0 before sum

# Save sum in time column
dt['time'] = total_min

# print update dataframe
dt
```

```python
#Now we divide stars column in two one for stars and the other one for ratings
dt['ratings'] = dt['stars']
dt
```

```python
#we keep just stars informations in stars column
dt['stars'] = dt['stars'].str.extract(r'(\d+\.\d+|\d+)')
dt['stars']
```

```python
#for Nan value we will change them by zero
dt['stars'].fillna(value=0.0, inplace=True)
dt['stars']
```

```python
#Now we select ratings in ratings's column
#we keep just stars informations in stars column
dt['ratings'] = dt['ratings'].str.extract(r'(\d+) ratings')
dt['ratings']
```

```python
#We will just replace Nan values by O
dt['ratings'].fillna(value=0.0, inplace=True)
dt['ratings']
```
